```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head(5)
```

```python
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year, df['per capita income (US$)'], color = 'red', marker='+')
plt.show()
```

```python
#fit the model
new_df = df.drop('per capita income (US$)', axis= 'columns')
new_df
```

```python
per_capita_income_usd = df['per capita income (US$)']
per_capita_income_usd
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,per_capita_income_usd)
```

```python
predict = reg.predict([[2020]])
predict
```

```python
plt.scatter(new_df.year, df['per capita income (US$)'], color='red', marker='+', label='Original Data')

# Plotting the predicted value
plt.plot([2020], predict, color='blue', marker='o', markersize=8, label='Predicted Value')

plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.legend()
plt.show()
```
